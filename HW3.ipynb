```python
import logomaker
import pandas as pd

# Define sequences
sequences = [
    "ACTGACTGACTG",
    "ACTAACTGACTC",
    "ACTGACTGACTG",
    "ACTGACTGACCC",
    "ACTGACTGACTG",
    "ACTGACTGACTA"
]

# Convert sequences to alignment matrix
counts_df = logomaker.alignment_to_matrix(sequences)

# Normalize the matrix to represent probabilities
prob_df = counts_df / counts_df.sum()

# Define widths
widths = pd.Series([1, 1, 1, 0.5, 0.5, 0.5, 1, 1, 1, 0.5, 0.5, 0.5], index=counts_df.index)

# Create sequence logo
logo = logomaker.Logo(prob_df, color_scheme='N_base', width=widths)

# Display sequence logo
logo.ax.set_xlabel('Position')
logo.ax.set_ylabel('Frequency')
plt.show()
```

```python
from Bio import SeqIO, motifs
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# Define sequences
sequences = [
    SeqRecord(Seq("ACTGACTGACTG"), id="seq1"),
    SeqRecord(Seq("ACTAACTGACTC"), id="seq2"),
    SeqRecord(Seq("ACTGACTGACTG"), id="seq3"),
    SeqRecord(Seq("ACTGACTGACCC"), id="seq4"),
    SeqRecord(Seq("ACTGACTGACTG"), id="seq5"),
    SeqRecord(Seq("ACTGACTGACTA"), id="seq6")
]

# Write sequences to a FASTA file
with open("sequences.fasta", "w") as f:
    SeqIO.write(sequences, f, "fasta")

# Create motif
m = motifs.create([record.seq for record in sequences])

# Generate sequence logo
m.weblogo("my_sequence_logo1.png")
```
